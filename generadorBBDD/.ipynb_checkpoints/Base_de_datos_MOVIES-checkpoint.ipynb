```python
df = pd.read_csv("Movies.csv", encoding='ISO-8859-1')
# Eliminar el año entre paréntesis al final del título
df["Movie or Series"] = df["Movie or Series"].str.replace(r"\s\(\d{4}\)$", "", regex=True)
df.to_csv("Movies_DEF.CSV", index=False)
```

```python
import csv

titulo = input("Por favor, inserte aquí el título de la película serie o saga: ")
year = input("Por favor, inserte aquí el año: ")
categoria = input ("Por favor, elija una de las categorías:\n A) COMEDIAS\n B) INTRIGA\n C) ACCIÓN\n D) MIEDO o THRILLER\n E) ROMÁNTICAS\n F) SAGAS y SERIES\n G)Investment Banking\n \nELIGE A, B, C, D, E, F o G.\n")
movie_nueva = []

while categoria not in ["A", "B", "C", "D", "E", "F", "G"]:
    print("Esta opción no es válida, por favor, prueba otra vez.")
    categoria = input ("Por favor, elija una de las categorías:\n A) COMEDIAS\n B) INTRIGA\n C) ACCIÓN\n D) MIEDO o THRILLER\n E) ROMÁNTICAS\n F) SAGAS y SERIES\n G)Investment Banking\n \nELIGE A, B, C, D, E, F o G.\n")
else:
    if (categoria == "A"):
        movie_nueva.append("COMEDIAS")
    elif (categoria == "B"):
        movie_nueva.append("INTRIGA")
    elif (categoria == "C"):
        movie_nueva.append("ACCIÓN")
    elif (categoria == "D"):
        movie_nueva.append("MIEDO o THRILLER")
    elif (categoria == "E"):
        movie_nueva.append("ROMÁNTICAS")
    elif (categoria == "F"):
        movie_nueva.append("SAGAS y SERIES")
    else:
        movie_nueva.append("Investment Banking")
    movie_nueva.append(titulo)
    movie_nueva.append(int(year))
    print(movie_nueva)
    
    # Abre el archivo CSV en modo de "append" ('a') para añadir sin sobrescribir
    with open('Movies_DEF.csv', mode='a', newline='') as archivo:
        writer = csv.writer(archivo)

        # Escribe la nueva fila en el archivo
        writer.writerow(movie_nueva)
```
