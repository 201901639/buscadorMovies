```python
import csv

titulo = input("Por favor, inserte aquí el título de la película serie o saga: ")
year = input("Por favor, inserte aquí el año: ")
categoria = input ("Por favor, elija una de las categorías:\n A) COMEDIES\n B) INTRIGUE\n C) ACTION\n D) THRILLER\n E) ROMANTIC\n F) SAGAS AND SERIES\n G)FINANCE\n \nELIGE A, B, C, D, E, F o G.\n")
movie_nueva = []

while categoria not in ["A", "B", "C", "D", "E", "F", "G"]:
    print("Esta opción no es válida, por favor, prueba otra vez.")
    categoria = input ("Por favor, elija una de las categorías:\n A) COMEDIES\n B) INTRIGUE\n C) ACTION\n D) THRILLER\n E) ROMANTIC\n F) SAGAS AND SERIES\n G)FINANCE\n \nELIGE A, B, C, D, E, F o G.\n")
else:
    if (categoria == "A"):
        movie_nueva.append("COMEDIES")
    elif (categoria == "B"):
        movie_nueva.append("INTRIGUE")
    elif (categoria == "C"):
        movie_nueva.append("ACTION")
    elif (categoria == "D"):
        movie_nueva.append("THRILLER")
    elif (categoria == "E"):
        movie_nueva.append("ROMANTIC")
    elif (categoria == "F"):
        movie_nueva.append("SAGAS AND SERIES")
    else:
        movie_nueva.append("FINANCE")
    movie_nueva.append(titulo)
    movie_nueva.append(int(year))
    print(movie_nueva)
    
    # Abre el archivo CSV en modo de "append" ('a') para añadir sin sobrescribir
    with open('generadorBBDD/Movies_DEF.csv', mode='a', newline='') as archivo:
        writer = csv.writer(archivo)

        # Escribe la nueva fila en el archivo
        writer.writerow(movie_nueva)
```
