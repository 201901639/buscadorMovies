```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer los datos con correccin de codificacin
movies_DEF = pd.read_csv("generadorBBDD/Movies_DEF.csv", encoding='utf-8')

# Limpieza bsica
movies_DEF.drop_duplicates(inplace=True)
movies_DEF.dropna(inplace=True)

# Correccin de texto (si es necesario)
movies_DEF['Movie or Series'] = movies_DEF['Movie or Series'].str.replace('', '').str.strip()

# Visualizacin: distribucin por categora
plt.figure(figsize=(12, 6))
sns.countplot(data=movies_DEF, y='Category', order=movies_DEF['Category'].value_counts().index, palette='viridis')
plt.title('Distribucin de Pelculas/Series por Categora', fontsize=16)
plt.xlabel('Cantidad')
plt.ylabel('Categora')
plt.show()

# Visualizacin: tendencias por ao
movies_DEF['Year'] = pd.to_numeric(movies_DEF['Year'], errors='coerce')  # Asegurar que 'Year' sea numrico
movies_DEF = movies_DEF[movies_DEF['Year'].notna()]  # Eliminar filas con aos invlidos

plt.figure(figsize=(12, 6))
sns.histplot(data=movies_DEF, x='Year', bins=20, kde=True, color='blue')
plt.title('Tendencias de Produccin por Ao', fontsize=16)
plt.xlabel('Ao')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Tabla estilizada: Top 10 pelculas/series ms recientes
movies_recent = movies_DEF.sort_values(by='Year', ascending=False).head(10)
movies_recent.style.background_gradient(cmap='coolwarm').set_properties(**{'text-align': 'left'})
```

```python
import pandas as pd
import streamlit as st

# Leer la base de datos
movies_DEF = pd.read_csv("generadorBBDD/Movies_DEF.csv", encoding='utf-8')

# Limpiar datos si es necesario
movies_DEF['Movie or Series'] = movies_DEF['Movie or Series'].str.replace('', '').str.strip()
movies_DEF['Category'] = movies_DEF['Category'].str.strip()

# Configurar la app de Streamlit
st.title("Buscador de Pelculas y Series")
st.subheader("Explora fcilmente las pelculas y series de la base de datos")

# Campo de bsqueda
search_term = st.text_input("Buscar por ttulo, categora o ao:")

# Filtrar datos segn el trmino de bsqueda
if search_term:
    filtered_data = movies_DEF[
        movies_DEF.apply(lambda row: search_term.lower() in str(row.values).lower(), axis=1)
    ]
    st.write(f"Se encontraron {len(filtered_data)} resultados:")
    st.dataframe(filtered_data)
else:
    st.write("Introduce un trmino de bsqueda para comenzar.")

# Mostrar todos los datos si no hay bsqueda activa
st.write("Base de datos completa:")
st.dataframe(movies_DEF)
```
